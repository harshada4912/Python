1. Given a list [4, 8, 2, 10, 6], find the maximum number using a for loop.

```python
nums = [4, 8, 2, 10, 6]
max_num = nums[0]

for n in nums:
    if n > max_num:
        max_num = n

print(max_num)
```

2. Take input from the user and find the factorial of the number using a for loop.

```python
n=int(input("Enter the number:"))
fact=1
for i in range(1, n + 1):
    fact *= i

print("Factorial =", fact)
```

4. Take a string input from the user and count the number of uppercase letters using a for loop. (Use char.isupper())

```python
text = input("Enter text: ")
count = 0

for ch in text:
    if ch.isupper():
        count += 1

print("Uppercase letters =", count)
A
```

10. Given a list of words ['Python', 'Java', 'C++', 'JavaScript', 'Ruby'], find the longest word using a for loop.

```python
words = ['Python', 'Java', 'C++', 'JavaScript', 'Ruby']
longest = words[0]

for w in words:
    if len(w) > len(longest):
        longest = w

print(longest)
```

11. Take a string input from the user and count the number of vowels using a for loop. (Vowels: a, e, i, o, u)

```python
text = input("Enter text: ")
vowels = "aeiouAEIOU"
count = 0

for ch in text:
    if ch in vowels:
        count += 1

print("Vowels =", count)
```

13. Given a list [1, 3, 5, 7, 9, 11], check if a user-inputted number is present in the list using a for loop.

```python
nums = [1, 3, 5, 7, 9, 11]
x = int(input("Enter number: "))

found = False
for n in nums:
    if n == x:
        found = True
        break

print("Present" if found else "Not present")
```
