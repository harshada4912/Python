### 4.Write a program to count vowels in a string.

```python
s = "harshada"
vowels = "aeiouAEIOU"
count = 0
for ch in s:
    if ch in vowels:
        count += 1
print(count)
```

### 5.Write a program to count consonants in a string.

```python
s = "Anuja"
vowels = "aeiouAEIOU"
count = 0
for ch in s:
    if ch.isalpha() and ch not in vowels:
        count += 1
print(count)
```

### 20.Write a program to swap the first and last character of a string.

```python
s = "python"
if len(s) > 1:
    s = s[-1] + s[1:-1] + s[0]
print(s)
```
