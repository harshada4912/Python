 Write a Reverse a list without using reverse()

```python
numbers = [1, 2, 3, 4, 5]
reversed_list = numbers[::-1]
print(reversed_list)
```

Write a Remove duplicates from a list

```python
nums = [1, 2, 2, 3, 4, 4, 5]

nums = list(set(nums))
print(nums)
```

Find the largest and smallest number

```python
nums = [10, 50, 3, 75, 20]
largest = max(nums)
smallest = min(nums)
print(largest, smallest)
```

Write a Count occurrences of an element

```python
nums = [1, 2, 2, 3, 2, 4]
nums.count(2)
```

Find even numbers from a list

```python
nums = [10, 3, 5, 8, 20, 7]
even = [x for x in nums if x % 2 == 0]
print(even)
```

Find the second largest number

```python
nums = [10, 50, 30, 20, 40]
nums.sort()
second_largest = nums[-2]
print(second_largest)
```

Print all elements of a list

```python
lst = [10, 20, 30]

for i in lst:
    print(i)
```

Find the length of a list

```python
lst = [1, 2, 3, 4]
len(lst)
```
