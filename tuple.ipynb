1. Write a program to create a tuple with numbers.

```python
t = (10, 20, 30, 40)
print(t)
```

2. Write a program to access elements from a tuple.


```python
t=(10, 20, 30, 40)
print(t[0])
print(t[1])
print(t[2])
print(t[3])
```

3. Write a program to print the first and last element of a tuple.

```python
t=(10, 20, 30, 40)
print("First:",t[0])
print("last:",t[-1])
```

4. Write a program to find the length of a tuple.

```python
t = (1, 2, 3, 4)
print(len(t))
```

5. Write a program to convert a list into a tuple.

```python
lst = [10, 20, 30]
t = tuple(lst)
print(t)
```

6. Write a program to check if an element exists in a tuple.



```python
t=(1,23,34,23,5)
if 23 in t:
    print("23 found")
else:
    print("23 not found")
```

7. Write a program to count how many times an element appears in a tuple.

```python
t=(1,2,3,3,4,5)

print(t.count(3))
```

9. Write a program to concatenate two tuples.

```python
a = (1, 2)
b = (3, 4)
c = a + b
print(c)
```

10. Write a program to repeat a tuple 3 times.


```python
t = (5, 10)
print(t * 3)
```
